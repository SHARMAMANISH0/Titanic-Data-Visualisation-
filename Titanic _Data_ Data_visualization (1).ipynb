```python
sns.countplot(x=manish ['Survived'])
```

```python
# If we want to  see acutal value than we use  value_count()..
manish['Survived'].value_counts()
```

```python
manish['Survived'].value_counts().plot(kind='bar')
```

```python
# If we want to check for pclass than....

sns.countplot(x=manish['Pclass'])
```

```python
# If we want to find  out percentage than use piechart
manish['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```
